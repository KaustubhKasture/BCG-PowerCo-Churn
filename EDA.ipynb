## 3. Data Visualisation

```python
def plot_stacked_bars(dataframe,title_,size_=(18,10),rot_=0,legend_='upper right'):
    ax = dataframe.plot(
        kind='bar',
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_)
    annotate_stacked_bars(ax,textsize=14)
    plt.legend(['Retension','Churn'],loc=legend_)
    plt.ylabel("Company Base (%)")
    plt.show

def annotate_stacked_bars(ax,pad=0.99,color='white',textsize=13):
    for p in ax.patches:
        value = str(round(p.get_height(),1))
        if value == 0.0:
            continue
        ax.annotate(value,
                   ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
                   color=color,
                   size=textsize)
```

```python
churn = client_df[['id','churn']]
churn.columns = ['Companies','churn']
total_churn = churn.groupby(churn['churn']).count()
churn_percentage = total_churn/total_churn.sum()*100
trans = churn_percentage.transpose()
```

```python
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

Above barplot shows about 10% of total customers have churned

### Sales Channel

```python
client_sales = client_df[['id','channel_sales','churn']]
client_sales = client_sales.groupby([client_sales['channel_sales'],client_sales['churn']])['id'].count().unstack().fillna(0)
client_sales_churn = (client_sales.div(client_sales.sum(axis=1),axis=0)*100).sort_values(by=[1],ascending=False)
```

```python
plot_stacked_bars(client_sales_churn, 'Sales channel', rot_=20)
```

Churning customers are distrubuted amoung 5 different `channel_sales`

## Consumption
Visualisation of distribution of consumption in last year and month

```python
consumption = client_df[['id','cons_12m','cons_gas_12m','cons_last_month','imp_cons','has_gas','churn']]
```

```python
fig,axs = plt.subplots(nrows=4,figsize=(18,25))

#Histogram
sns.boxplot(consumption['cons_12m'],ax=axs[0])
sns.boxplot(consumption[consumption['has_gas'] == 't']['cons_gas_12m'],ax=axs[1])
sns.boxplot(consumption['cons_last_month'],ax=axs[2])
sns.boxplot(consumption['imp_cons'],ax=axs[3])

#Adjusting axis
for ax in axs:
    ax.ticklabel_format(style='plain',axis='x')
    axs[0].set_xlim(-200000, 2000000)
    axs[1].set_xlim(-200000, 2000000)
    axs[2].set_xlim(-20000, 350000)
plt.show()
```

The consumption data is highly positively skewed with a very considerable tail towards higher values of distribution

### Forecast

```python
forecast = client_df[['id', 'forecast_cons_12m','forecast_cons_year','forecast_discount_energy','forecast_meter_rent_12m',
          'forecast_price_energy_off_peak','forecast_price_energy_peak','forecast_price_pow_off_peak','churn']]
```

```python
def plot_distribution(dataframe,column,ax,bins_=50,rot_=0):
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
                         "Churn":dataframe[dataframe["churn"]==1][column]})
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True,rot=rot_)
    ax.set_xlabel(column)
    ax.ticklabel_format(style='plain', axis='x')    
```

```python
fig,axs = plt.subplots(nrows=7,figsize=(18,50))

#Histograms
plot_distribution(client_df, "forecast_cons_12m", axs[0])
plot_distribution(client_df, "forecast_cons_year", axs[1])
plot_distribution(client_df, "forecast_discount_energy", axs[2])
plot_distribution(client_df, "forecast_meter_rent_12m", axs[3])
plot_distribution(client_df, "forecast_price_energy_off_peak", axs[4])
plot_distribution(client_df, "forecast_price_energy_peak", axs[5])
plot_distribution(client_df, "forecast_price_pow_off_peak", axs[6])
```

Similar to consumption, forecast data is positively skewed and has a long tail on side of higher values of distribution
